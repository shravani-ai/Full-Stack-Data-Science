```python
#Fibonacci Series
n = 10
a = 0
b = 1
next = b
count = 1

while count <= n:
    print(next, end=" ")
    count += 1
    a, b = b, next
    next = a + b
print()
```

```python
#Factorial number
def fact(n):
  f = 1
  for i in range(1, n+1):
    f = f*i
  return f
x = 5
result = fact(x)
print(result)
```

```python
#Factorial number using Recursion(Function calling itself)
def fact(n):
  if n==0:
    return 1
  return n * fact(n-1)
result = fact(5)
result

```
