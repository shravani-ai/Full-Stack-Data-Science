```python
from sklearn.neighbors import KNeighborsRegressor
knn_model = KNeighborsRegressor()
knn_model.fit(x,y)
```

Tree Algorithm

1. Decision Tree 

```python
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor()
dt_model.fit(x,y)
```

```python
dt_model_pred = dt_model.predict([[6.5]])
print(dt_model_pred)
```

```python
dt_model = DecisionTreeRegressor(criterion='squared_error',splitter='best')
dt_model.fit(x,y)
dt_model_pred = dt_model.predict([[6.5]])
print(dt_model_pred)
```

```python
dt_model = DecisionTreeRegressor(criterion='squared_error',splitter='random')
dt_model.fit(x,y)
dt_model_pred = dt_model.predict([[6.5]])
print(dt_model_pred)
```

```python
dt_model = DecisionTreeRegressor(criterion='absolute_error',splitter='best')
dt_model.fit(x,y)
dt_model_pred = dt_model.predict([[6.5]])
print(dt_model_pred)
```

```python
dt_model = DecisionTreeRegressor(criterion='absolute_error',splitter='random')
dt_model.fit(x,y)
dt_model_pred = dt_model.predict([[6.5]])
print(dt_model_pred)
```
