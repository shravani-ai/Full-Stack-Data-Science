```python
'''polynomial == poly + nomial (increase degree of indepedent variable)
'''
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\omkar joshi\DATA SCIENCE\Databases\emp_sal.csv")
df
```

```python
x = df.iloc[:,1:2].values
y = df.iloc[:,2].values
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x,y)
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg.predict(x),color = 'blue')
plt.title("Linear Regression Graph")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures()
x_poly = poly_reg.fit_transform(x)
```

```python
x_poly #model of one degree
```

```python
poly_reg.fit(x_poly,y)
lin_reg2 = LinearRegression() #Linear model build with 2nd degree
lin_reg2.fit(x_poly,y)
```

Polynomial Model

```python
plt.scatter(x,y, color = 'red')
plt.plot(x, lin_reg2.predict(poly_reg.fit_transform(x)), color ="blue")
plt.title("Truth or Bluff (Polynomial Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

```python
lin_model_pred = lin_reg.predict([[6.5]])
print(lin_model_pred)

poly_model_pred = lin_reg2.predict(poly_reg.fit_transform([[6.5]]))
print(poly_model_pred)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=3)
x_poly = poly_reg.fit_transform(x)
poly_reg.fit(x_poly,y)
lin_reg2 = LinearRegression() #Linear model build with 4th degree
lin_reg2.fit(x_poly,y)
plt.scatter(x,y, color = 'red')
plt.plot(x, lin_reg2.predict(poly_reg.fit_transform(x)), color ="blue")
plt.title("Truth or Bluff (Polynomial Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

```python
poly_model_pred = lin_reg2.predict(poly_reg.fit_transform([[6.5]]))
print(poly_model_pred)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
x_poly = poly_reg.fit_transform(x)
poly_reg.fit(x_poly,y)
lin_reg2 = LinearRegression() #Linear model build with 4th degree
lin_reg2.fit(x_poly,y)
plt.scatter(x,y, color = 'red')
plt.plot(x, lin_reg2.predict(poly_reg.fit_transform(x)), color ="blue")
plt.title("Truth or Bluff (Polynomial Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

```python
poly_model_pred = lin_reg2.predict(poly_reg.fit_transform([[6.5]]))
print(poly_model_pred)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=5)
x_poly = poly_reg.fit_transform(x)
poly_reg.fit(x_poly,y)
lin_reg2 = LinearRegression() #Linear model build with 4th degree
lin_reg2.fit(x_poly,y)
plt.scatter(x,y, color = 'red')
plt.plot(x, lin_reg2.predict(poly_reg.fit_transform(x)), color ="blue")
plt.title("Truth or Bluff (Polynomial Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

SVR (Support vector Regression)

```python
'''svr(dataset in contineous varible(regression))
svm(dataset is binary(classification))

best fit line in simple linear regression --> hyperplane in svr
The distance between teo support line should always maximum.
(you adjust the error and accuracy will increase)

The distance between two support line is called marginal line.

by default degree is 3 in svr
by default degree is 2 in polynomial

polynomia predict with 5th degree -- 6.5 00 174.8
svr model predict with (karner=poly, degree = 4, gamma = ' auto, C= 10.0 ) -- 6.5 svr - 175.7

model ranking table 
1.svr -- 6.5 -> 175.8
2.KNN --6.5 -> 175.3
3.poly -- 6.5 -> 174.8
4.RFR -- 6.5 -> 168
5.DTR -- 6.5 -> 150
'''
```

```python
from sklearn.svm import SVR
svr_model = SVR()
svr_model.fit(x,y)
```

```python
svr_model_pred = svr_model.predict([[6.5]])
print(svr_model_pred)
```

```python
svr_model = SVR(kernel='sigmoid', degree=4, gamma='auto',C= 10.0)
svr_model.fit(x,y)
svr_model_pred = svr_model.predict([[6.5]])
print(svr_model_pred)
```

```python
svr_model = SVR(kernel='poly', degree=4, gamma='auto',C= 10.0)
svr_model.fit(x,y)
svr_model_pred = svr_model.predict([[6.5]])
print(svr_model_pred)
```

```python
svr_model = SVR(kernel='linear', degree=4, gamma='auto',C= 10.0)
svr_model.fit(x,y)
svr_model_pred = svr_model.predict([[6.5]])
print(svr_model_pred)
```

```python
svr_model = SVR(kernel='sigmoid', degree=4, gamma='scale',C= 10.0)
svr_model.fit(x,y)
svr_model_pred = svr_model.predict([[6.5]])
print(svr_model_pred)
```

```python
svr_model = SVR(kernel='sigmoid', degree=3, gamma='auto',C= 15.0)
svr_model.fit(x,y)
svr_model_pred = svr_model.predict([[6.5]])
print(svr_model_pred)
```
